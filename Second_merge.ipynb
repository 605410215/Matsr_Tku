## Import Library

```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

## Load song_json 

```python
df2=pd.read_csv('song.json',sep='\t')
df2.rename(columns={'id': 'song_id'}, inplace=True)
#df2['artist_id']=df2['artist_id'].astype(str)
df2.head()
```

## Load Various_Artist

```python
Various_Artist=pd.read_csv('Various_Artist',sep='\t')
Various_Artist.head()
```

## Merge artist by song_id and Remove Various_Artist

```python
size=0
start_time=tm.time()
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='a'+chr(ii)+chr(j)
        df=pd.read_csv(file_name,sep='\t')
        df=df.merge(df2,on='song_id', how='left')
        df.drop(['album_id'], axis=1, inplace=True)
        df=df[~df['artist_id'].isin(Various_Artist.artist_id)]
        df.to_hdf(file_name+'.h5','a')
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
end_time=tm.time()-start_time
print ('Data_size:%d Process_time:%d'%(size,end_time))
```
