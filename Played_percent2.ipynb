```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

```python
size=0
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='a'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        df['hear_gap']=df.hear_percent.apply(lambda a:1 if a<=0.5 else 2 if (a>0.5) & (a<=1) else 3 )
        df.to_hdf(file_name,'a')
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break

print ('Data_size:%d'%(size))
```
