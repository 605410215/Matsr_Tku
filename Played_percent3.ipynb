```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

```python
size=0
start_time=tm.time()
columns=['msno','song_id','song_length','played_length','day','artist_id','hear_percent','hear_gap']
dataset2=pd.DataFrame(columns=columns)
dataset2=dataset2.groupby(['msno','artist_id']).count()
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='a'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        df=df.groupby(['msno','artist_id']).count()
        dataset2=dataset2.add(df,fill_value=0)
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
end_time=tm.time()-start_time
print ('Data_size:%d Process_time:%d'%(size,end_time))
```

```python
dataset3=dataset2.reset_index()
dataset3.head()
```

```python
dataset3=dataset3.drop(['song_length','played_length','day','hear_percent','hear_gap'],axis=1)
dataset3.rename(columns={'song_id': 'number'}, inplace=True)
dataset3.head()
```

```python
dataset3.to_hdf('msno_artist.h5','a')
```
