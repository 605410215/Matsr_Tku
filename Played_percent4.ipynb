```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

```python
size=0
columns=['msno','song_id','song_length']
dataset2=pd.DataFrame(columns=columns)
dataset2=dataset2.groupby(['msno','song_id']).count()
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='a'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        df=df.drop(['played_length','day','artist_id','hear_percent','hear_gap'],axis=1)
        df=df.groupby(['msno','song_id']).count()
        dataset2=dataset2.add(df,fill_value=0)
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
print ('Data_size:%d'%(size))
```
