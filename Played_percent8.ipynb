```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
import itertools
```

```python
df=pd.read_hdf('msno_artist_percent.h5','a')
df=df[df['SP']==2]
df.head(100)
```

```python
df.shape
```

```python
T=df.groupby('artist_id')['sum_number'].sum().reset_index().sort_values('sum_number',ascending=False)
T=T.reset_index(drop=True)
T
```

```python
#T.to_csv('A.csv',sep=',',encoding='utf-8',index=False)
```

```python
T['Q']=T.index
T
```

```python
T['W']=T['Q'].apply(lambda n : 1 if n<23019 else 2 if n<50000 else 3 if n< 80000 else 4)
T
```

```python
T[T['W']==4]
```

```python
S=T.groupby(['W'])['sum_number'].sum().reset_index()
S
```

```python
R=T.groupby(['W']).count().reset_index().drop('Q',axis=1)
R
```

```python
R=R.drop('sum_number',axis=1)
E=R.merge( S, on='W', how='left')
E
```

```python
E.to_csv('Ab.csv',sep=',',encoding='utf-8',index=False)
```

```python
E['F']=np.log10(E.artist_id)
E['FF']=np.log10(E.sum_number)
E
```
