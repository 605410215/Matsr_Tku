```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
import itertools
```

```python
df=pd.read_hdf('msno_artist_percent.h5','a')
df.head()
```

```python
df.number.sum()
```

```python
df=df[df['SP']==2]
df=df[df['percent']>=0.6]
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
W=df.drop_duplicates(subset='msno', keep='first', inplace=False)['msno'].values
W
```

```python
artist_dic=dict()
def calculate_Dic(artist_key):
    if artist_key in artist_dic:
        artist_dic[artist_key]=artist_dic[artist_key]+1
    else:
        artist_dic[artist_key]=1
    return artist_dic
```

```python
def combine_artist(pd_msno):
    print(pd_msno)
    return list(map(calculate_Dic,list(itertools.combinations(df[df['msno']==pd_msno]['artist_id'], 2))))
```

```python
list(map(combine_artist,W))
```
