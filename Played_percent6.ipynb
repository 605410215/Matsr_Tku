```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

```python
df=pd.read_hdf('sum_hear_gap.h5','a')
df['artist_id']=df['artist_id'].astype(str)
df['artist_id']=df['artist_id'].apply(lambda n:n[:n.index('.')])
df['artist_id']=df['artist_id'].astype('int64')
df
```

```python
df.dtypes
```

```python
df['SP']=df.hear_gap.apply(lambda n : 1 if n<2 else 2)
#df['mach']=df['msno'].astype(str)+'_'+df['artist_id'].astype(str)+'_'+df['SP'].astype(str)
df.head(10)
```

```python
df.dtypes
```

```python
df2=df.groupby(['msno','artist_id',"SP"])['number'].sum().reset_index()
df2.head()
```

```python
df2.dtypes
```

```python
df3=df2.groupby(['msno','artist_id'])['number'].sum().reset_index()
df3.rename(columns={'number': 'sum_number'}, inplace=True)
df3.head()
```

```python
df3.dtypes
```

```python
df4=df2.merge(df3, on=['msno','artist_id'], how='left')
df4.head()
```

```python
df4['percent']=df4.number/df4.sum_number
df4.head(10)
```
